 ## Subtask 2.4

In mass spring damper system a spring force makes the mass move. Let's write it with physical formulas by using Newton's law.

Let x be the position of mass at time t.

m - the mass.

c - viscous damping coefficient.

k - spring constant.

u - force input.



$m\ddot{x}+c\dot{x}+kx=u$

Let's substitute new variables.

$\ddot{x} = \dot{x_2}$

$\dot{x} = \dot{x_1}=x_2$

$x = x_1$

So, to state-space:

$\dot{x} = \begin{pmatrix}0&1 \\
-\frac{k}{m}&-\frac{c}{m}
\end{pmatrix}*x + \begin{pmatrix}
0 \\
\frac{u}{m}
\end{pmatrix}$

I put random values for $k, m, c$ and $u$.
In the code below you can see my simulation.


```
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from math import sin

n = 2 # order of the equation
A = np.array([[0, 1],
              [-30, -1]])

# x_dot from state space
def StateSpace(x, t):
    return A.dot(x)+np.array([0, 10])

# time horizon
time = np.linspace(0, 1, 100)
x0 = np.random.rand(n)  # initial state

# the solution in the a dictionary form
solution = {"SS": odeint(StateSpace, x0, time)}

# plt.subplot(121)
plt.plot(time, solution["SS"])
plt.xlabel('time')
plt.ylabel('x(t)')

plt.show()
```
