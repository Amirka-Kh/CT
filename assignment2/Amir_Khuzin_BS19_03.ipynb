## Task 1.1 Make the following systems stable, proposing appropriate control
####1) 
We have a system: $$\dot x=\begin{pmatrix} 10 & 0 \\ -5 & 10 \end{pmatrix}x+\begin{pmatrix} 2 \\ 0 \end{pmatrix}u$$
For the following system we should propose such control, that will make the system stable.
In this example $A$ matrix is not stable since its eigenvalues are positive $l_1=10$ and $l_2=10$, therefore we should find appropriate control $u$.
1. To find $u$ we will use the Pole placement method. Poles should be non-positive, let us assume that they are equal to $[-1, -2]$.
2. Now, let us define $u=-[k_1, k_2]x$. 
3. Substituting $u$ to the system gives us closed-loop system $\dot x=(A-BK)x$.
4. Let us express eigenvalues for the closed-loop system: 
$$(A-BK)=\begin{pmatrix} 10-2k_1 & -2k_2 \\ -5 & 10 \end{pmatrix}$$ 
$$\begin{pmatrix} 10 -2k_1-s & -2k_2 \\ -5 & 10-s \end{pmatrix}$$
$$(10-2k_1-s)(10-s)-10k_2=$$
$$s^2-(20-2k_1)s+(100-20k_1-10k_2)=0$$
5. Remember the poles $l_1=-1$ and $l_2=-2$. We will find $k_1$ and $k_2$ values by comparing coefficients of $s$ terms. 
$$(s+1)(s+2)=$$
$$s^2+3s+2=$$
$$s^2-(20-2k_1)s+(100-20k_1-10k_2)=0$$
$$\begin{bmatrix} 2k_1-20=3 \\ 100-20k_1-10k_2=2 \end{bmatrix}$$
So, we get $k_1=11.5$ and $k_2=-13.2$
6. Finally, $u=-[11.5, -13.2]x$.

## Task 2.1 Plot root locus

**For a system with $A$ with imaginary eigenvalues**

Let $A=\begin{pmatrix} 0 & -1 \\ 1 & 0 \end{pmatrix}$ whose eigenvalues are $[j, -j]$

```
import matplotlib.pyplot as plt
from numpy.linalg import eig
from scipy.integrate import odeint
import numpy as np

A = np.array([[0, -1], [1, 0]])
B = np.array([[1], [1]])
K0 = np.array([[1, 2]]);

k_min = 1;
k_max = 10;
k_step = 0.1;

Count = np.floor((k_max-k_min)/k_step)
Count = Count.astype(int)

k_range = np.linspace(k_min, k_max, Count)

E = np.zeros((Count, 4))

def get_eigen(K0):
  for i in range(Count):
    K0[0, 1] = k_range[i]

    ei, v = eig((A - B.dot(K0)))

    E[i, 0] = np.real(ei[0])
    E[i, 1] = np.imag(ei[0])
    E[i, 2] = np.real(ei[1])
    E[i, 3] = np.imag(ei[1])

    #print("eigenvalues of A - B*K:", ei)
def plot_graph(E):
  plt.plot(E[:, 0], E[:, 1])
  plt.plot(E[:, 2], E[:, 3])
  plt.xlabel('real')
  plt.ylabel('imag')
  plt.show()
get_eigen(K0)  
plot_graph(E)  
```

#Task 5.1
We have a system:
$$\dot{x}=\begin{pmatrix}10&0\\-5&10\end{pmatrix}x+\begin{pmatrix}2\\0\end{pmatrix}u$$
let us design point-to-point control for it. 

Firstly, let us drive it toward the point $x^*=\begin{pmatrix}0\\1\end{pmatrix}$

Define control as: $u=-K(x-x^*)+u^*$

Okey, we know that $\dot{x^*}=0$ and that at the node our dynamics obtains the form: $0=\begin{pmatrix}10&0\\-5&10\end{pmatrix}
\begin{pmatrix}0\\1\end{pmatrix}+
\begin{pmatrix}2\\0\end{pmatrix}u^*$ 
In other words, $u^*=0$.

We define control error 'e' and write error dynamics:
$$\dot{x}-\dot{x^*}=\begin{pmatrix}10&0\\-5&10\end{pmatrix}x- 
\begin{pmatrix}10&0\\-5&10\end{pmatrix}x^*+
\begin{pmatrix}2\\0\end{pmatrix}
(-K(x-x^*)+u^*)-\begin{pmatrix}2\\0
\end{pmatrix}u^*$$


$$
\dot{e}= 
\begin{pmatrix}10&0\\-5&10\end{pmatrix}e
-
\begin{pmatrix}2\\0\end{pmatrix}Ke
$$

```
from scipy.signal import place_poles
from scipy.integrate import odeint
import numpy as np
from numpy.linalg import eig

n = 2

#feedback design
A = np.array([[10, 0], [-5, 10]])
B = np.array([[2], [0]])

#desired eigenvalues
poles = np.array([-1, -2])
place_obj = place_poles(A, B, poles)

#found control gains
K = place_obj.gain_matrix;
print("K:", K)

#test that eigenvalues of the closed loop system are what they are supposed to be 
e, v = eig((A - B.dot(K)))
print("eigenvalues of A - B*K:", e)
```

Simulate forward wirth the found control

```
from matplotlib import pyplot as plt

x_desired = np.array([0, 1])
u_desired = np.array([0])


def StateSpace(x, t):
    u = -K.dot(x - x_desired) + u_desired
    return A.dot(x) + B.dot(u)

time = np.linspace(0, 30, 30000)   
x0 = np.random.rand(n)  # initial state

solution = {"solution_1": odeint(StateSpace, x0, time)}    


plt.rcParams['figure.figsize'] = [20, 20]

# plt.rcParams["font.family"] = "Old Times American"
plt.rcParams["font.size"] = 16
plt.rcParams["font.weight"] = 'normal'


# plt.subplot(221)
plt.plot(time, solution["solution_1"], linewidth=2)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.grid(color='k', linestyle='--', linewidth=0.7, alpha=0.3)
# plt.title('autonomous')
```
